# 웹페이지 크롤링
- 웹페이지에서 원하는 값만 가져오는 것 -> 크롤링-스크래이핑
- 파이썬에서는 BeautifulSoup로 대표되는 다양한 크롤링 툴이 존재
    * BS4만으로 크롤링 하는 것은 입문 수준이므로, 실제 복잡하고 정교한 툴도 존재할 수 있음
- 실제 크롤링의 세계는 별도의 학계체계, 규칙 등이 존재하므로 영리를 위한 스크래이핑이면 철저한 사전조사와 자문이 필요(저작권 이슈, 초상권 등 개인정보 문제 발생할 수 있음)

## 01. HTML
- HTML은 웹사이트에서 표준으로 사용된고 있는 마크업 언어(문서가 화면에 표시되는 형식 및 논리적 구조를 느슨하게 설명한 체계)
- "<>" + "</>" 두가지 기호로 구분되어짐

## 02. url
- 한편, 웹페이지의 주소는 일정한 체계를 가지고 있음
- UCI : Universial Content Identifier, 자료에 고유 번호체계가 존재함, 이 규칙만 뽑아낼 수 있다면 이 또한 가능
- http://www.uci.or.kr/?menuno=18 (자세한 내용은 좀 더 찾아봐야 합니다)

```
https://archive.aks.ac.kr/insp/popupTextComp.do?dataUCI=G002+AKS+KSM-XE.1429.4717-20101008.B001a_001_00289_XXX
```

여튼, 현 시점에서는 해당 웹페이지의 자료를 추출하는 것을 목적으로 합니다.

````
from bs4 import BeautifulSoup
import requests
````

````
source = "https://archive.aks.ac.kr/insp/popupTextComp.do?dataUCI=G002+AKS+KSM-XB.0000.0000-20101008.B004a_004_00297_YYY"
````

````
response = requests.get(source)
````

````
if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.select_one('#wrap > div > section > div > div > dl > dd > div > div')
    #print(title)
    print(title.get_text())
else : 
    print(response.status_code)
````

### source를 리스트로 만들어 여러 파일을 저장해도 되고, UCI규칙을 발견한다면 **반복문(for)**을 활용해서 여러 페이지를 한번에 출력하고 저장할 수도 있다..!

## 03. file write&read
- 추출한 텍스트를 일단 파일로 만들어 저장하고 이를 불러올 필요가 있다..!
- 파이썬에는 fopen으로 이를 수행
- r(읽기)/w(쓰기)/a(추가)


```
파일 객체 = open(파일 이름, 파일 열기 모드)
```




````
f = open("sample.txt", 'w')
f.write(title.get_text())
f.close()
````

````
# readline_all.py
f = open("sample.txt", 'r')
while True:
    line = f.readline()
    if not line: break
    print(line)
f.close()
````
